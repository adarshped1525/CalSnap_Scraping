#### Using BeautifulSoup 

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.canada.ca/en/health-canada/services/food-nutrition/healthy-eating/nutrient-data/nutrient-value-some-common-foods-2008.html#a8"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

**Vegetables Data**

```python
table = soup.find_all('table')[3]
print(table)
```

```python
nutrition_titles  = table.find_all('th')
nutrition_titles
```

```python
nutrition_table_titles = [title.text for title in nutrition_titles[:22]]

print(nutrition_table_titles)
```

```python
import pandas as pd

df_Vegetables  = pd.DataFrame(columns = nutrition_table_titles)
df_Vegetables
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[3:-12]:
    row_data = row.find_all('td')
    individual_row_data_Vegetables = [data.text for data in row_data[:22]]
    #print(individual_row_data_Vegetables)
    
    length = len(df_Vegetables)
    df_Vegetables.loc[length] = individual_row_data_Vegetables
```
